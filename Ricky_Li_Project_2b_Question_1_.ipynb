```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Import the required libraries

# Storing the dataframe from the URL
student_data = pd.read_csv('https://raw.githubusercontent.com/rickyli8529/Data-Analytics-Datasets/main/StressLevelDataset.csv')

# Plotting a correlation matrix of the dataset
plt.figure(figsize=(20, 8))
sns.heatmap(student_data.corr(), annot=True, fmt ='.0%')

# Creating lists for each factors in the dataset
physiological_factors = ['headache', 'blood_pressure', 'sleep_quality', 'breathing_problem']
academic_factors = ['academic_performance', 'study_load', 'teacher_student_relationship', 'future_career_concerns']
psychological_factors = ['anxiety_level', 'self_esteem', 'depression']

# Plotting multiple pie charts of each category in physiological factors list using a for loop to go through each factor and plot each one
for factor in physiological_factors:
    factor_data = student_data[factor].value_counts() # Counting values of each factor
    plt.figure(figsize=(5, 10))   # Setting the size of the plot
    factor_data.plot.pie(title=f'Pie Chart of Student {factor}') # Plotting the pie chart for one factor
    plt.show()  # Showing the plot

# Plotting multiple horizontal bar graphs of each category in academic factors list using a for loop to go through each factor and plot each one
for factor in academic_factors:
  factor_data = student_data[factor].value_counts() # Counting values of each factor
  plt.figure(figsize = (10, 5)) # Setting the size of the plot
  sns.barplot(x = factor_data.values, y = factor_data.index, alpha = 1, orient = 'h') # Plotting the bar graph for one factor
  plt.title(f'Distribution of Student {factor}')  # Creating the title of the chart
  plt.ylabel(f'{factor} Level', fontsize = 12) # Creating the y label of the chart
  plt.xlabel('Number of Students', fontsize = 12) # Creating the x label of the chart
  plt.show()  # Showing the plot

# Plotting multiple vertical bar graphs of each category in psychological factors list using a for loop to go through each factor and plot each one
for factor in psychological_factors:
  factor_data = student_data[factor].value_counts() # Counting values of each factor
  plt.figure(figsize = (15, 10))  # Setting the size of the plot
  sns.barplot(x = factor_data.index, y = factor_data.values, alpha = 0.5) # Plotting the bar graph for one factor
  plt.title(f'Distribution of {factor} of Students')  # Creating the title of the chart
  plt.ylabel('Number of Students', fontsize = 12) # Creating the y label of the chart
  plt.xlabel(f'{factor} Level', fontsize = 12)  # Creating the x label of the chart
  plt.show()  # Showing the plot
```
